Your task is to code up the algorithm from the video lectures for computing strongly connected components (SCCs), and to run this algorithm on the given graph.

Output Format: You should output the sizes of the 5 largest SCCs in the given graph, in decreasing order of sizes, separated by commas (avoid any spaces)

```python
import numpy as np
import sys
sys.setrecursionlimit(10 ** 6)

def DFS_Loop(G) :

  DFS_Loop.t = 1
  visited = [False] * 875715
  t_values = [0] * 875715
  Stack = []

  groups = [0] * 875715
  DFS_Loop.s = 0

  for i in range(875714, 0, -1):
    print('This is {}th vertex'.format(i))
    if not visited[i]:
      DFS_Loop.s = i
      DFS(G, i, visited, Stack, groups, t_values)

      
  return (visited, t_values, groups, Stack)

def DFS(G, v, visited, stack, groups, t_values) :# G : graph, v : start vertex, visited : stack
  if visited[v] == True:
    pass
  else:
    stack.append(v)
    visited[v] = True
    groups[v] = DFS_Loop.s
   
  print(f'Current vertex is {v} and the t-value is {DFS_Loop.t}')
  
  if False not in np.array(visited)[sorted(G[v])]: # if every adjecent node is already visited
    t_values[v] = DFS_Loop.t
    DFS_Loop.t += 1  ################ does this affect the t value in DFS_Loop? #############################
    stack.pop()
    if stack == []: # if stack is empty
      pass
    else:
      DFS(G, stack[-1], visited, stack, groups, t_values) 

  else:
    for i in sorted(G[v]): # Traverse from small one to big one # in ascending order
      if not visited[i]:
        DFS(G, i, visited, stack, groups, t_values)
  
```
